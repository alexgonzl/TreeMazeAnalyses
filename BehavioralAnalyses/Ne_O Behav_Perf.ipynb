```python
%matplotlib notebook

import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
file_name = '~/Google Drive/PostDoc/Data/SwitchBehavioralPerformance.csv'
df = pd.read_csv(file_name)
```

```python
subj_name = 'O'

f = plt.figure()
ax1 = plt.subplot(211)
plt.plot(df[subj_name + '-TC'],'o-')
plt.plot(df[subj_name + '-SC'],'o-')
plt.axhspan(0,50,facecolor='0.8',alpha=0.7)
plt.grid(True)
plt.setp(ax1.get_xticklabels(),visible=False)
plt.ylim(0,100)
plt.xlim(-1,30)
plt.title(subj_name+' Performance')
plt.legend(('% Correct','% Switch Cor.'),loc='lower right')

ax1 = plt.subplot(212)
subj_name = 'Ne'
plt.plot(df[subj_name + '-TC'],'o-')
plt.plot(df[subj_name + '-SC'],'o-')
plt.axhspan(0,50,facecolor='0.8',alpha=0.7)
plt.grid(True)
plt.ylim(0,100)

plt.xlim(-1,30)
plt.title(subj_name+' Performance')

plt.show()


f.savefig("Ne_O_BehavPerf.pdf",bbox_inches='tight')
```
